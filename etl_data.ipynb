```python
import requests
import pandas as pd
```

```python
# URL do endpoint
url = "https://api.obrasgov.gestao.gov.br/obrasgov/api/projeto-investimento"
```

```python
# Parametros iniciais
params = {
    "uf": "DF",
    "page": 0,
    "size": 10  # limite max, testado usando script
}
```

```python
all_data = []

while True:
    response = requests.get(url, params=params)
    if response.status_code != 200:
        print("Erro: ", response.status_code, response.text)
        break
    
    data = response.json()
    content = data.get("content", [])
    
    if not content:
        break 
    
    all_data.extend(content)
    
    # Vai para proxima pagina caso nao seja a ultima
    if data.get("last", True):
        break
    params["page"] += 1
```

```python
# Transformando tudo em DataFrame
df = pd.json_normalize(all_data)
print(df.head())
print("Total de registros: ", len(df))
```
